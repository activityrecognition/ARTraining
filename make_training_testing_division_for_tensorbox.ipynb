```python
import json, os

tensorbox_input_filepath = "../ARThermal/tensorbox_dataset/tensorbox_input.json"
training_json_path = "../ARThermal/tensorbox_dataset/train_boxes.json"
testing_json_path = "../ARThermal/tensorbox_dataset/val_boxes.json"
```

```python
with open(tensorbox_input_filepath, "r") as tb_in:
    tensorbox_input = json.load(tb_in)
    
with open("../ARThermal/tensorbox_dataset_poses/tensorbox_input.json", "r") as tb_poses:
    tensorbox_input.extend(json.load(tb_poses))
```

```python
for entry in tensorbox_input:
    entry["image_path"] = "/workspace" + entry["image_path"][2:]
```

```python
len(tensorbox_input)
```

```python
import random

data_len = len(tensorbox_input)

videos = {}
for i in tensorbox_input:
    video_id = os.path.basename(os.path.dirname(i["image_path"]))
    if videos.get(video_id):
        videos[video_id].append(i)
    else:
        videos[video_id] = [i]

print len(videos.keys())
training_count = 0
training_dataset = []
while len(videos.keys()) > 0: 
    if len(training_dataset) >= int(data_len*0.8):
        break
    
    k = random.choice(videos.keys())
    #print k
    training_dataset.extend(videos[k])
    #print videos[k]
    del videos[k]

testing_dataset = [i for i in tensorbox_input if i not in training_dataset]
random.shuffle(testing_dataset)
```

```python
print len(training_dataset), float(len(training_dataset))/data_len
print len(testing_dataset), float(len(testing_dataset))/data_len
```

```python
with open(training_json_path, "w") as f:
    json.dump(training_dataset, f, indent=2)
    
with open(testing_json_path, "w") as f:
    json.dump(testing_dataset, f, indent=2)
```
